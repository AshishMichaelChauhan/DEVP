```python
# Descriptive analysis for non-categorical data - Measures of Central Tendency

# Select non-categorical (numerical) columns
numerical_cols = df_uq.select_dtypes(include=['number'])
def cen_ten_mes(df_uq):
    measures = pd.DataFrame({
        'Minimum': df_uq.min(),
        'Maximum': df_uq.max(),
        'Mean': df_uq.mean().round(2),
        'Median': df_uq.median(),
        'Mode': df_uq.mode().iloc[0], # Mode may have multiple values, take the first one
        '25th Percentile': df_uq.apply(lambda x: np.percentile(x.dropna(), 25)),
        '50th Percentile': df_uq.apply(lambda x: np.percentile(x.dropna(), 50)),
        '75th Percentile': df_uq.apply(lambda x: np.percentile(x.dropna(), 75))
    })
    return measures

# Calculate and display the measures of central tendency for numerical columns
measures = cen_ten_mes(numerical_cols)
print(measures)
```

```python
# Descriptive analysis for non-categorical data - Measures of Dispersion

# Import relevant libraries
import scipy.stats as stats
from scipy.stats import skew, kurtosis

# Selecting non-categorical (numeric) columns
numeric_df = df_uq.select_dtypes(include=[np.number])

# Function to calculate measures of dispersion
def calculate_dispersion(df_uq):
    dispersion_metrics = {}

    # Range
    dispersion_metrics['Range'] = df_uq.max() - df_uq.min()

    # Standard Deviation
    dispersion_metrics['Standard Deviation'] = df_uq.std().round(2)

    # Skewness
    dispersion_metrics['Skewness'] = df_uq.apply(lambda x: skew(x.dropna())).round(4)

    # Kurtosis
    dispersion_metrics['Kurtosis'] = df_uq.apply(lambda x: kurtosis(x.dropna(), fisher=False)).round(4)

    return pd.DataFrame(dispersion_metrics)

# Correlation matrix
correlation_matrix = numeric_df.corr().round(4)

# Calculate measures of dispersion
dispersion_results = calculate_dispersion(numeric_df)

# Display results
print("Dispersion Measures:")
print(dispersion_results)

print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
## Descriptive analysis for non-categorical data - Composite Measure
# Selecting non-categorical (numeric) columns
numeric_df = df_uq.select_dtypes(include=[np.number])

# Function to calculate Coefficient of Variation (CV)
def coefficient_of_variation(df_uq):
    return (df_uq.std() / df_uq.mean()) * 100
# Function to calculate 95% Confidence Interval
def confidence_interval(df_uq, confidence=0.95):
    ci_dict = {}
    for col in df_uq.columns:
        sample_data = df_uq[col].dropna()  # Remove NaN values
        mean = np.mean(sample_data)
        sem = stats.sem(sample_data)  # Standard Error of the Mean
        margin = sem * stats.t.ppf((1 + confidence) / 2., len(sample_data) - 1)
        ci_dict[col] = (round(mean - margin, 4), round( mean + margin, 4))
    return ci_dict

# Calculating Coefficient of Variation
cv_results = numeric_df.apply(coefficient_of_variation).round(4)

# Calculating Confidence Interval
ci_results = confidence_interval(numeric_df)


# Display Results
print("Coefficient of Variation (in %):")
print(cv_results)

print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci}")
```

```python
# Infrential Analysis - Test of Mean {t | ANOVA}
#Import relevant libraries
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Selecting non-categorical (numeric) columns
numeric_df = df_uq.select_dtypes(include=[np.number])

# T-Test Function (Assuming two-sample independent t-test)
def t_test(data, col1, col2):
    t_stat, p_val = stats.ttest_ind(df_uq['Quantity'].dropna(), df_uq['Value'].dropna())
    return {'T-Statistic': round(t_stat, 4), 'P-Value': round(p_val, 4)}

# Define the ANOVA test function
def anova_test(df_uq, dependent_var, independent_var):
    model = ols(f'{dependent_var} ~ C({independent_var})', data=df_uq).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    return anova_table.round(4)

t_test_result = t_test(numeric_df, 'Quantity', 'Value') 

anova_results = anova_test(df_uq, 'Value', 'Category')

# Display results
print("T-Test Result:")
print(t_test_result)

print("\nANOVA Result:")
print(anova_results)
```

```python
#Infrential Analysis - Test of Variance {Levene}

# Group data by Shipping Method
air_quantity = df_uq[df_uq['Shipping_Method'] == 'Air']['Quantity']
sea_quantity = df_uq[df_uq['Shipping_Method'] == 'Sea']['Quantity']
land_quantity = df_uq[df_uq['Shipping_Method'] == 'Land']['Quantity']

# Perform Levene's test to check for equality of variance across Shipping Methods
stat, p_value = stats.levene(air_quantity, sea_quantity, land_quantity)

# Print results
print(f"Levene’s test statistic: {stat}")
print(f"p-value: {p_value}")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Variances are significantly different (reject null hypothesis).")
else:
    print("Variances are equal (fail to reject null hypothesis).")
```

```python
# Infrential Analysis - Test of Proportion {z | Chi-sq}
# Convert 'Quantity' to categorical based on median
df_uq['Quantity_Category'] = pd.cut(df_uq['Quantity'], bins=[-np.inf, df_uq['Quantity'].median(), np.inf], labels=['Below Median', 'Above Median'])

# Convert 'Value' to categorical based on median
df_uq['Value_Category'] = pd.cut(df_uq['Value'], bins=[-np.inf, df_uq['Value'].median(), np.inf], labels=['Below Median', 'Above Median'])

# Create a contingency table for Chi-square test
contingency_table = pd.crosstab(df_uq['Quantity_Category'], df_uq['Value_Category'])
print("Contingency Table:")
print(contingency_table)

# Chi-Square Test of Independence
chi2, p_chi2, _, _ = stats.chi2_contingency(contingency_table)
print(f"\nChi-Square Test:")
print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p_chi2}")

# For z-test, we'll use proportions from categorical data
# Example: Testing proportions of 'Above Median' for 'Quantity'
n = len(df_uq)
p_observed = (df_uq['Quantity_Category'] == 'Above Median').mean()
p_expected = 0.5  # Hypothetical expected proportion

# z-test for proportions
z = (p_observed - p_expected) / np.sqrt((p_expected * (1 - p_expected)) / n)
p_z = stats.norm.cdf(z)
print(f"\nZ-Test:")
print(f"Z Statistic: {z}")
print(f"P-value: {p_z}")
```

```python
#Infrential Analysis - Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera}

#import relevant libraries
from scipy.stats import shapiro, kstest, anderson, jarque_bera

#Apply normality tests to Quantity, Value, and Weight

# 1. Shapiro-Wilk Test
print("Shapiro-Wilk Test:")
for column in ['Quantity', 'Value', 'Weight']:
    stat, p_value = shapiro(df_uq[column])
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")

# 2. Kolmogorov-Smirnov Test
# Assuming normal distribution parameters for ks test
print("\nKolmogorov-Smirnov Test:")
for column in ['Quantity', 'Value', 'Weight']:
    mean = np.mean(df_uq[column])
    std = np.std(df_uq[column], ddof=1)
    stat, p_value = kstest(df[column], 'norm', args=(mean, std))
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")

# 3. Anderson-Darling Test
print("\nAnderson-Darling Test:")
for column in ['Quantity', 'Value', 'Weight']:
    result = anderson(df_uq[column], dist='norm')
    print(f"{column}: statistic={round(result.statistic, 4)}, critical values={result.critical_values}, significance levels={result.significance_level}")

# 4. Jarque-Bera Test
print("\nJarque-Bera Test:")
for column in ['Quantity', 'Value', 'Weight']:
    stat, p_value = jarque_bera(df_uq[column])
    print(f"{column}: statistic={round(stat, 4)}, p-value={round(p_value, 4)}")
```

```python
#Infrential Analysis - Non-Parametric Tests {Mann-Whitney | Wilcoxon | Kruskal-Wallis | Friedman} 

# Import relevant libraries
from scipy.stats import mannwhitneyu, wilcoxon, kruskal, friedmanchisquare


# Convert 'Date' to ordinal (numeric format) for comparison
df_uq['Date'] = pd.to_datetime(df_uq['Date'], errors='coerce')
df_uq['Date'] = df_uq['Date'].apply(lambda x: x.toordinal() if pd.notnull(x) else np.nan)

# Round numeric columns to 4 decimal places
df_uq[['Quantity', 'Value', 'Weight']] = df_uq[['Quantity', 'Value', 'Weight']].round(4)

# Mann-Whitney U Test (comparing two independent variables)
stat, p_value = mannwhitneyu(df_uq['Quantity'], df_uq['Value'])
print(f"Mann-Whitney U Test - Statistic: {stat:.4f}, P-value: {p_value:.4f}")

# Wilcoxon Signed-Rank Test (comparing paired samples)
stat, p_value = wilcoxon(df_uq['Quantity'], df_uq['Value'])
print(f"Wilcoxon Signed-Rank Test - Statistic: {stat:.4f}, P-value: {p_value:.4f}")

# Kruskal-Wallis H Test (comparing three or more independent variables)
stat, p_value = kruskal(df_uq['Quantity'], df_uq['Value'], df_uq['Weight'])
print(f"Kruskal-Wallis H Test - Statistic: {stat:.4f}, P-value: {p_value:.4f}")

# Friedman Test (comparing three or more paired samples)
stat, p_value = friedmanchisquare(df_uq['Quantity'], df_uq['Value'], df_uq['Weight'])
print(f"Friedman Test - Statistic: {stat:.4f}, P-value: {p_value:.4f}")
```

```python
# 1. Test of Mean (t-Test)
#Hypothesis:
#Null Hypothesis (H₀): There is no significant difference in the average value of transactions between exports and imports.
#Alternative Hypothesis (H₁): There is a significant difference in the average value of transactions between exports and imports.
#python

from scipy.stats import ttest_ind

# Define the groups
export_values = df[df['Import_Export'] == 'Export']['Value']
import_values = df[df['Import_Export'] == 'Import']['Value']

# Perform the t-test
t_stat, p_value = ttest_ind(export_values, import_values)

# Conclusion
if p_value < 0.05:
    print(f"Reject H₀. Significant difference in means. t-stat: {t_stat}, p-value: {p_value}")
else:
    print(f"Fail to reject H₀. No significant difference in means. t-stat: {t_stat}, p-value: {p_value}")
```

```python
# 2. ANOVA (Analysis of Variance)
#Hypothesis:
#Null Hypothesis (H₀): The means of transaction values across different countries (e.g., Colombia, Chile, Turkey) are equal.
#Alternative Hypothesis (H₁): At least one country has a significantly different mean transaction value.
#python

from scipy.stats import f_oneway

# Group data by country
colombia = df[df['Country'] == 'Colombia']['Value']
chile = df[df['Country'] == 'Chile']['Value']
turkey = df[df['Country'] == 'Turkey']['Value']

# Perform ANOVA
f_stat, p_value = f_oneway(colombia, chile, turkey)

# Conclusion
if p_value < 0.05:
    print(f"Reject H₀. At least one country has a different mean. F-stat: {f_stat}, p-value: {p_value}")
else:
    print(f"Fail to reject H₀. All countries have similar means. F-stat: {f_stat}, p-value: {p_value}")
```

```python
# 3. Test of Variance (Levene's Test & Bartlett's Test)
#Hypothesis (Levene's Test):
#Null Hypothesis (H₀): The variances of transaction values across different countries (e.g., Colombia, Chile, Turkey) are equal.
#Alternative Hypothesis (H₁): The variances of transaction values across different countries are not equal.
#python

from scipy.stats import levene

# Perform Levene's test
lev_stat, lev_p = levene(colombia, chile, turkey)

# Conclusion
if lev_p < 0.05:
    print(f"Reject H₀. The variances are not equal. Levene's stat: {lev_stat}, p-value: {lev_p}")
else:
    print(f"Fail to reject H₀. The variances are equal. Levene's stat: {lev_stat}, p-value: {lev_p}")
```

```python
# 4. Test of Proportion (Chi-Square Test)
#Hypothesis:
#Null Hypothesis (H₀): There is no significant association between the type of transaction (Import/Export) and payment terms.
#Alternative Hypothesis (H₁): There is a significant association between the type of transaction (Import/Export) and payment terms.
#python

from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['Import_Export'], df['Payment_Terms'])

# Perform Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Conclusion
if p < 0.05:
    print(f"Reject H₀. Significant association between Import/Export and Payment Terms. Chi2: {chi2}, p-value: {p}")
else:
    print(f"Fail to reject H₀. No significant association between Import/Export and Payment Terms. Chi2: {chi2}, p-value: {p}")
```

```python
# 5. Test of Correlation (Pearson’s Correlation)
#Hypothesis:
#Null Hypothesis (H₀): There is no significant linear correlation between transaction quantity and value.
#Alternative Hypothesis (H₁): There is a significant linear correlation between transaction quantity and value.
#python

from scipy.stats import pearsonr

# Perform Pearson's correlation
corr_stat, p_value = pearsonr(df['Quantity'], df['Value'])

# Conclusion
if p_value < 0.05:
    print(f"Reject H₀. There is a significant correlation. Corr stat: {corr_stat}, p-value: {p_value}")
else:
    print(f"Fail to reject H₀. No significant correlation. Corr stat: {corr_stat}, p-value: {p_value}")
```

```python
# 6. Non-Parametric Test (Mann-Whitney U Test)
#Hypothesis:
#Null Hypothesis (H₀): There is no significant difference in the distribution of transaction values between export and import.
#Alternative Hypothesis (H₁): There is a significant difference in the distribution of transaction values between export and import.
#python

from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test
stat, p_value = mannwhitneyu(export_values, import_values)

# Conclusion
if p_value < 0.05:
    print(f"Reject H₀. Significant difference in the distribution of transaction values. Stat: {stat}, p-value: {p_value}")
else:
    print(f"Fail to reject H₀. No significant difference in the distribution of transaction values. Stat: {stat}, p-value: {p_value}")
```

```python
# Categorical Variables - Chi-sq Test {Goodness of Fit | Test of Independence}

# Initialize lists to store results\

variable_list = []
test_list = []
chi2_statistic_list = []
p_value_list = []

for var in categorical_vars:
    # Goodness of Fit Test
    observed = df_uq[var].value_counts()
    expected = [len(df_uq) / len(observed)] * len(observed)  # Equal expected frequencies
    chi2_stat, p_val = stats.chisquare(observed, expected)
    
    # Append Goodness of Fit results
    variable_list.append(var)
    test_list.append('Goodness of Fit')
    chi2_statistic_list.append(chi2_stat)
    p_value_list.append(p_val)
    
    # Test of Independence (cross-tabulation with itself for simplicity here)
    contingency_table = pd.crosstab(df_uq[var], df_uq[var])
    chi2_stat, p_val, _, _ = stats.chi2_contingency(contingency_table)
    
    # Append Test of Independence results
    variable_list.append(var)
    test_list.append('Test of Independence')
    chi2_statistic_list.append(chi2_stat)
    p_value_list.append(p_val)

# Convert results to DataFrame
results_df = pd.DataFrame({
    'Variable': variable_list,
    'Test': test_list,
    'Chi2 Statistic': chi2_statistic_list,
    'P-Value': p_value_list
})

# Display results
print(results_df)
```

```python
# Categorical Variables - Non-Parametric Tests {Mann-Whitney | Wilcoxon | Kruskal-Wallis | Friedman}

# Assuming 'Numeric_Var' is the numeric variable we are testing against categorical variables
numeric_var = 'Value'

# Initialize lists to store results
variable_list = []
test_list = []
statistic_list = []
p_value_list = []

# Mann-Whitney U Test (for two independent groups)
for var in categorical_vars:
    categories = df_uq[var].unique()
    if len(categories) == 2:  # Mann-Whitney U Test requires exactly two groups
        group1 = df_uq[df_uq[var] == categories[0]][numeric_var]
        group2 = df_uq[df_uq[var] == categories[1]][numeric_var]
        u_stat, p_val = stats.mannwhitneyu(group1, group2)
        variable_list.append(var)
        test_list.append('Mann-Whitney U')
        statistic_list.append(u_stat)
        p_value_list.append(p_val)

# Wilcoxon Signed-Rank Test (requires paired samples)
for var in categorical_vars:
    categories = df_uq[var].unique()
    if len(categories) == 2:  # Wilcoxon Signed-Rank Test also requires exactly two groups
        group1 = df_uq[df_uq[var] == categories[0]][numeric_var]
        group2 = df_uq[df_uq[var] == categories[1]][numeric_var]
        # Ensure both groups have same number of samples
        if len(group1) == len(group2):
            w_stat, p_val = stats.wilcoxon(group1, group2)
            variable_list.append(var)
            test_list.append('Wilcoxon Signed-Rank')
            statistic_list.append(w_stat)
            p_value_list.append(p_val)

# Kruskal-Wallis H Test (for more than two independent groups)
for var in categorical_vars:
    categories = df_uq[var].unique()
    if len(categories) > 2:  # Kruskal-Wallis H Test requires more than two groups
        groups = [df_uq[df_uq[var] == category][numeric_var] for category in categories]
        h_stat, p_val = stats.kruskal(*groups)
        variable_list.append(var)
        test_list.append('Kruskal-Wallis H')
        statistic_list.append(h_stat)
        p_value_list.append(p_val)

# Friedman Test (for more than two related groups)
# Here we assume 'Numeric_Var' is repeated for different conditions within the same subjects
for var in categorical_vars:
    categories = df_uq[var].unique()
    if len(categories) > 2:  # Friedman Test requires more than two related groups
        # Create a DataFrame suitable for Friedman test
        data = [df_uq[df_uq[var] == category][numeric_var].values for category in categories]
        if all(len(d) == len(data[0]) for d in data):  # Ensure all groups have the same number of samples
            f_stat, p_val = stats.friedmanchisquare(*data)
            variable_list.append(var)
            test_list.append('Friedman')
            statistic_list.append(f_stat)
            p_value_list.append(p_val)

# Convert results to DataFrame
results_df = pd.DataFrame({
    'Variable': variable_list,
    'Test': test_list,
    'Statistic': statistic_list,
    'P-Value': p_value_list
})

# Display results
print(results_df)
```
